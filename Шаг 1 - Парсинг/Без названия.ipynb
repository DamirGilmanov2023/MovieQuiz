```python
from bs4 import BeautifulSoup
import requests
from PIL import Image
from io import BytesIO
```

```python
url="https://fonwall.ru/search?q=кадры из фильма&page=4&format=js&seed=2023-09-15T04:09:46.887Z"
```

```python
response=requests.get(url)
text=response.text
print(text)
```

```python
encoded_text = text.encode('utf-8').decode('windows-1251')
print(encoded_text)
```

```python
text=text.split(";")
print(text[1])
```

```python
html=text[1][43:-2]
mass=html.split("><")
```

```python
img_urls=[]
for m in mass:
    if m[:4]=="img ":
        src=m.split("src=")
        src=src[1].split('\"')
        src=src[1]
        img_urls.append(src)
```

```python
i=1
for urls in img_urls:
    response = requests.get(urls)
    if response.status_code == 200:
        image_data = response.content
        img = Image.open(BytesIO(image_data))
        img.save(f'{i}.jpg')
        print("Изображение успешно загружено и сохранено.")
        i+=1
    else:
        print(f"Ошибка при загрузке изображения. Код статуса: {response.status_code}")
```

```python
i=1
iterator=0
while True:
    iterator+=1
    print(f"Загрузка {iterator} страницы")
    try:
        url=f"https://fonwall.ru/search?q=кадры из фильма&page={iterator}&format=js&seed=2023-09-15T04:09:46.887Z"
        response=requests.get(url)
        response.encoding = "utf-8"
        text=response.text
        text=text.split(";")
        html=text[1][43:-2]
        mass=html.split("><")
        img_urls=[]
        img_alts=[]
        for m in mass:
            if m[:4]=="img ":
                src=m.split("src=")
                src=src[1].split('\"')
                src=src[1]
                img_urls.append(src)
                alt=m.split("alt=")
                alt=alt[1].split('\"')
                alt=alt[1]
                img_alts.append(alt)
        for urls,alts in zip(img_urls,img_alts):
            response = requests.get(urls)
            if response.status_code == 200:
                image_data = response.content
                img = Image.open(BytesIO(image_data))
                img.save(f'{i}.jpg')
                with open('urls.txt', 'a') as file:
                    text_to_add = f"{i} - {urls}"
                    file.write(text_to_add + '\n')
                with open('alts.txt', 'a') as file:
                    text_to_add = f'{i}:\"{alts}\"'
                    file.write(text_to_add + ',\n')
                print("Изображение успешно загружено и сохранено.")
                i+=1
            else:
                print(f"Ошибка при загрузке изображения. Код статуса: {response.status_code}")
    except:
        print("Вероятно конец")
```
